```python
import os
import numpy as np
from PIL import Image
import matplotlib.pylab as plt
```

```python
rootDir = 'photos'
array = np.zeros((20,200,200,3))
array1 = []
for dirName,subdirList,fileList in os.walk(rootDir):
    print("Found Directory: %s" % dirName)
    for fname in fileList:
        file_path = os.path.splitext(fname)
        outfile = file_path[0] + "_200x200" + file_path[1]
        if file_path[1] == ".jpg":
            new_width = 200
            new_height = 200
            im1 = Image.open(fname)
            im2 = im1.resize((new_width, new_height), Image.ANTIALIAS)
            im2.save(outfile,"JPEG")
            im3 = plt.imread(outfile)
            array1.append(im3)
        elif file_path[1] == ".png":
            new_width = 200
            new_height = 200
            im1 = Image.open(fname)
            im2 = im1.resize((new_width, new_height), Image.ANTIALIAS)
            im2.save(outfile,"PNG")
            im3 = plt.imread(outfile)
            array1.append(im3)
        for i in range(0,len(array1)):
            array[i] = array1[i]
print(array)
```
